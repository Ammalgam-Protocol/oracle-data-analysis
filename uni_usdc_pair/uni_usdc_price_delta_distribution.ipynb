```python
import pandas as pd
import plotly.express as px

# Read the CSV file
df = pd.read_csv('./data/uni_usdc_06132024.csv')

# Sort the dataframe by 'block_number'
df.sort_values(by='block_number', inplace=True)

# Naming 
# pair0 = UNI, pair1 = USDC => (UNI,USDC)

# Calculate the discrete relative change for 'pair0_cost'
df['discrete_relative_change_UNI'] = df['pair0_cost'].diff() / df['pair0_cost'].shift(1)

# Calculate the discrete relative change for 'pair1_cost'
df['discrete_relative_change_USDC'] = df['pair1_cost'].diff() / df['pair1_cost'].shift(1)
# Display the dataframe


#df.head()

# Drop NaN values resulting from the diff and shift operations
df.dropna(inplace=True)

# Plot interactive distribution graphs
fig_uni = px.histogram(df, x='discrete_relative_change_UNI', title='Distribution of Discrete Relative Change - UNI', marginal='box', nbins=50)
fig_usdc = px.histogram(df, x='discrete_relative_change_USDC', title='Distribution of Discrete Relative Change - USDC', marginal='box', nbins=50)

# Display the graphs
fig_uni.show()
fig_usdc.show()
```

```python
import numpy as np


# Calculate the continuous relative change for 'pair0_cost'
df['continuous_relative_change_UNI'] = np.log(df['pair0_cost'] / df['pair0_cost'].shift(1))

# Calculate the continuous relative change for 'pair1_cost'
df['continuous_relative_change_USDC'] = np.log(df['pair1_cost'] / df['pair1_cost'].shift(1))

# Drop NaN values resulting from the diff and shift operations
df.dropna(inplace=True)

# Plot interactive distribution graphs for continuous relative changes
fig_continuous_uni = px.histogram(df, x='continuous_relative_change_UNI', title='Distribution of Continuous Relative Change - UNI', marginal='box', nbins=50)
fig_continuous_usdc = px.histogram(df, x='continuous_relative_change_USDC', title='Distribution of Continuous Relative Change - USDC', marginal='box', nbins=50)

# Display the graphs

fig_continuous_uni.show()
fig_continuous_usdc.show()
```
